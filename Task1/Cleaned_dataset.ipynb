# DataSet: Netflix Movies and TV Shows

## 1. Load the Dataset

```python
import pandas as pd
df=pd.read_csv("netflix_titles.csv")
```

## 2. Inspect the Dataset

```python
df.info()
```

```python
df.head()
```

```python
df.describe(include="all")
```

## 3. Handle Missing Values

```python
#view missing values
df.isnull().sum()
```

```python
df['director'].fillna("Unknown",inplace=True)
```

```python
df['cast'].fillna("Not Available",inplace=True)
```

```python
df['country'].fillna("Unknown",inplace=True)
```

```python
df.dropna(subset=['date_added','rating'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['duration'],inplace=True)
```

## 4.Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

## 5. Standardize Text Columns


```python
df.head(3)
```

```python

df['type'] = df['type'].str.lower().str.strip()
df['rating'] = df['rating'].str.upper().str.strip()
df['country'] = df['country'].str.strip().str.title()
df['genre'] = df['listed_in'].str.lower().str.strip()  
```

## 6. Convert Date Formats

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

## 7. Rename Columns for Consistency

```python
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
```

## 8. Check and Convert Data Types

```python
# Make sure data types are appropriate
df['release_year'] = df['release_year'].astype(int)
df['date_added'] = pd.to_datetime(df['date_added'])
```
